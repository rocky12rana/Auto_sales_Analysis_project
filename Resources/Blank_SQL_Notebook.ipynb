```python
%%sql
SELECT*,
  quantity*netprice*exchangerate AS net_revenue
FROM
  sales
WHERE orderdate :: DATE >='2020-01-01'
LIMIT 10;
```

```python
%%sql
SELECT
  customer.customerkey,
  customer.givenname,
  customer.surname,
  customer.age,
  customer.countryfull,
  customer.continent,
  product.productkey,
  product.productname,
  product.categoryname,
  product.subcategoryname,
  sales.orderdate,
  quantity*netprice*exchangerate AS net_revenue
FROM
  sales
LEFT JOIN customer
ON sales.customerkey=customer.customerkey
LEFT JOIN product
ON sales.productkey=product.productkey
WHERE orderdate :: DATE >='2020-01-01'
LIMIT 10;
```
